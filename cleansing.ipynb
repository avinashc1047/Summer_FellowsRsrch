```python
human_labeled = []

# Open the spreadsheet and worksheet
worksheet = gc.open('dataset').sheet1 # replace 'My cool spreadsheet' with your sheet name

# Specify the column you want to get (e.g., 'A')
column_name = 'Self-Labeled'

cell = worksheet.find(column_name)
if cell is not None:
    # Get all values in the specified column
    column_values = worksheet.col_values(cell.col)

    # Remove the first value (column name) using slicing
    human_labeled = column_values[1:]

    print(human_labeled)
else:
    print(f"Column '{column_name}' not found in the worksheet.")
```

```python
def coversion(given):
  cnvrtd = []
  for i in range(len(given)):
      if given[i].lower() in ["related", "relevant", "1"]:
          cnvrtd.append(1)
      else: # This will handle "Irrelevant" and "Unrelated" cases
          cnvrtd.append(0)
  return cnvrtd
```

```python
human_rel = coversion(human_labeled) #my relevancy ratings
print(human_rel)

gemini_rel = coversion(gemini_labeled)
print(gemini_rel)
lambda_rel = coversion(lambda_labeled)
print(lambda_rel)
gpt_rel = coversion(gpt_labeled)
print(gpt_rel)
print(claude_labeled) # already 1 and 0
```

```python
# prompt: now i want to wrtie this information into the apporporate columns
import time  # Import the time module for sleep functionality

# Assuming 'human_rel', 'gemini_rel', 'lambda_rel', 'gpt_rel', and 'claude_labeled'
# are lists of the same length and correspond to the rows in your spreadsheet

# Open the spreadsheet and worksheet
worksheet = gc.open('dataset').sheet1  # Replace with your actual sheet name

# Get the number of rows in the worksheet
num_rows = len(worksheet.get_all_values())

# Start writing from the second row (index 1) to avoid overwriting headers
row_index = 2

for human, gemini, lambd, gpt, claude in zip(human_rel, gemini_rel, lambda_rel, gpt_rel, claude_labeled):
  worksheet.update_cell(row_index, 1, human)
  worksheet.update_cell(row_index, 2, gemini)
  worksheet.update_cell(row_index, 4, lambd)
  worksheet.update_cell(row_index, 3, gpt)
  worksheet.update_cell(row_index, 5, claude)
  row_index += 1
  time.sleep(1)  # Pause for 1 second after each write operation

print("Data written to Google Sheet successfully!")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

llm_stats = {}
```

```python
true_pos = 0
true_neg = 0
false_pos = 0
false_neg = 0

for index in range(len(human_rel)):
  if human_rel[index] == 1:
    true_pos += 1
  else:
    true_neg += 1

# Open the spreadsheet and worksheet
# Open the spreadsheet
spreadsheet = gc.open('dataset')

# Select the second worksheet by index (0-based indexing)
worksheet = spreadsheet.get_worksheet(1)  # This assumes 'sheet2' is the second sheet

worksheet.update_cell(4, 3, true_pos)
worksheet.update_cell(4, 4, false_neg)
worksheet.update_cell(5, 3, false_pos)
worksheet.update_cell(5, 4, true_neg)

# Assuming you have your true labels (y_true) and predicted labels (y_pred)
accuracy = accuracy_score(human_rel, human_rel)
precision = precision_score(human_rel, human_rel)
recall = recall_score(human_rel, human_rel)
f1 = f1_score(human_rel, human_rel)

llm_stats["Human"] = {
    "Accuracy": accuracy,
    "Precision": precision,
    "Recall": recall,
    "F1-score": f1
}

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1)

worksheet.update_cell(3, 7, f1)
worksheet.update_cell(4, 7, precision)
worksheet.update_cell(5, 7, recall)
worksheet.update_cell(3, 9, accuracy)

print("Data written to Google Sheet successfully!")
```

```python
f1_scores = []
accuracy_scores = []
precision_scores = []
recall_scores = []

for llm, stats in llm_stats.items():
    f1_scores.append(stats["F1-score"])
    accuracy_scores.append(stats["Accuracy"])
    precision_scores.append(stats["Precision"])
    recall_scores.append(stats["Recall"])

print(f1_scores)
print(accuracy_scores)
print(precision_scores)
print(recall_scores)
```

```python
plt.figure(figsize=(8, 6))
llm_names = list(llm_stats.keys())  # Extract names from the llm_stats dictionary
plt.xticks(range(len(llm_stats)), llm_names)
plt.plot(f1_scores, label="F1 Score")
plt.plot(accuracy_scores, label="Accuracy")
plt.plot(precision_scores, label="Precision")
plt.plot(recall_scores, label="Recall")
plt.xlabel("LLM Model")
plt.ylabel("Score")
plt.title("Comparison of LLM Performance Metrics")
plt.legend()
plt.grid(True)
plt.show()
```
